```python
carsalesmissing=pd.read_csv("carsales_missingdata.csv")
carsalesmissing
```

```python
#To fill the missing values with some value, lets say mean
carsalesmissing.Odometer.fillna(carsalesmissing.Odometer.mean())
```

```python
carsalesmissing
```

```python
# The values didn't change because pandas requires manual reassingnment of values in columns
# We could do something like: carsalesmissing.Odometer=carsalesmissing.Odometer.fillna(carsalesmissing.Odometer.mean())
# But we can do this more easily for filling NA values using inplace parameter
```

```python
carsalesmissing.Odometer.fillna(carsalesmissing.Odometer.mean(),
                               inplace=True)
carsalesmissing
```
