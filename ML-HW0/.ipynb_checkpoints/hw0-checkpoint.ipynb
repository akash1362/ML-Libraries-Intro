## Reading the dataset
We read the data into a pandas dataframe.
(We have to read from a local copy as the original data has an error in the 50th row)

Additionally, we can drop the Group and Name columns as the names have been removed and the group has to be ignored. After reading the dataset, we will display its first 5 data points.

```python
import pandas as pd
data = pd.read_csv('echocardiogram.data', header=None)
data.columns = ['Survival', 'Still alive', 'Age at Heart Attack', 'Pericardial Effusion',
                'Fractional Shortening', 'EPSS', 'LVDD', 'Wall Motion Score',
                'Wall Motion Index', 'Mult', 'Name', 'Group', 'Alive-at-1']

data = data.drop(['Group', 'Name'],axis=1)
print('Number of instances = %d' % (data.shape[0]))
print('Number of attributes = %d' % (data.shape[1]))
data.head()
```

## Missing Values
The missing values in the original dataset are encoded as '?'. Firstly, we replace these '?' with NaN and then count the number of missing values in each column.

```python
import numpy as np

data = data.replace('?', np.NaN)
print('Number of missing values:')
for col in data.columns:
    print('%s: %d' % (col,data[col].isna().sum()))

data[50:55]
```

Since all the columns contain missing values, which were represented with a '?', the values in all the columns are stored as string objects. Thus, we must convert all columns to numeric values.

```python
for col in data.columns:
    data[col] = pd.to_numeric(data[col])
```

Now, we will try different methods to deal with missing values.
### 1. Discard data points with missing values

```python
print("Number of rows in original data:", data.shape[0])
data2 = data.dropna()
print("Number of rows after dropping missing values:", data2.shape[0])
```


### 2. Imputation with median

```python
from sklearn.impute import SimpleImputer
imp = SimpleImputer(missing_values=np.nan, strategy='median')

data3 = data.copy(deep=True)
data3[data3.columns] = imp.fit_transform(data3[data3.columns])
data3[50:55]
```

### 3. Imputation with mean

```python
imp = SimpleImputer(missing_values=np.nan, strategy='mean')

data4 = data.copy(deep=True)
data4[data4.columns] = imp.fit_transform(data4[data4.columns])
data4[50:55]
```

## Standardization and Normalization


### Standardization
Standardization refers to the scaling of data so that all the data conforms to the same format.

We can standardize the data using the Z-score method.

```python
Z = (data-data.mean())/data.std()
Z.head()
```

We can also use this method to deal with outliers by discarding the instances with abnormally values for attributes in comparision to the Z-score.

```python
print('Number of rows before discarding outliers = %d' % (Z.shape[0]))

Z2 = Z.loc[((Z > -3).sum(axis=1)==11) & ((Z <= 3).sum(axis=1)==11),:]
print('Number of rows after discarding outliers = %d' % (Z2.shape[0]))
Z2.head()
```

Another approach to standardize our data is to scale our data so that it lies in the range [0, 1]. This can be done using **<font color="blue">MinMaxScaler</font>** from sklearn. 

```python
from sklearn import preprocessing
scaler = preprocessing.MinMaxScaler()
data_st1 = data.copy(deep=True)
data_st1[data_st1.columns] = scaler.fit_transform(data_st1[data_st1.columns])
data_st1.head()
```

We can also map our data to a normal distribution using **<font color="blue">QuantileTransformer</font>**

```python
quantile_transformer = preprocessing.QuantileTransformer(output_distribution='normal', random_state=0)
data_st2 = data.copy(deep=True)
data_st2[data_st2.columns] = quantile_transformer.fit_transform(data_st2[data_st2.columns])
data_st2.head()
```

### Normalization
Normalization is the process of scaling individual samples to have unit norm.
We can use the **<font color="blue">normalize</font>** function from sklearn to normalize our dataset.
(**<font color="blue">normalize</font>** does not allow NaN values, so we have to use a dataset where we have dealt with missing values)

```python
data_norm = data4.copy(deep=True);
data_norm[data_norm.columns] = preprocessing.normalize(data_norm[data_norm.columns], norm='l2')
data_norm.head()
```

## Discretization
Discretization provides a way to partition continuous features into discrete values. There are several methods to discretize the attributes in our dataset.

Firstly, let's plot a histogram that shows the distribution of the attribute values:

```python
data['LVDD'].hist(bins=10)
data['LVDD'].value_counts(sort=False)
```

### 1. Equal width
In this method, we apply the cut() function to discretize the attribute into 4 bins of similar interval widths.

```python
bins = pd.cut(data['LVDD'],4)
bins.value_counts(sort=False)
```

### 2. Equal frequency
For this method, we use the qcut() function to partition the values into 4 bins such that each bin has nearly the same number of instances.

```python
bins = pd.qcut(data['LVDD'],4)
bins.value_counts(sort=False)
```

### 3. Discretization using k-means clustering
We can use **<font color='blue'>KBinsDiscretizer</font>** to discretize our data using k-means clustering.

```python
from sklearn.preprocessing import KBinsDiscretizer

data5 = data2.copy(deep=True)
disc = KBinsDiscretizer(n_bins = 10, encode = 'ordinal', strategy = 'kmeans')
data5[['LVDD']] = disc.fit_transform(data5[['LVDD']])
data5['LVDD'].value_counts()
```

## Sampling
Sampling is the technique of selecting and analyzing a representative subset of data points from the dataset.
We can use various techniques to sample our data.

### 1. Sampling without replacement
In this method, we randomly select some samples from the dataset so that no sample can be chosen more than once.

We can randomly select a specific number of samples.

```python
sample = data.sample(n=3)
sample
```

Alternatively, we can select a specific percentage of out dataset (for example, 1% of the dataset)

```python
sample = data.sample(frac=0.01, random_state=58)
sample
```

### 2. Sampling with replacement

In this method, we randomly select some samples so that some of them might be repeated.

```python
sample = data.sample(frac=0.1, replace=True, random_state=58)
sample.head()
```

## Dimesionality Reduction
Dimensionality reduction is the transformation of data from a high-dimensional space into a low-dimensional space so that the low-dimensional representation retains some meaningful properties of the original data.

### 1. Principal Component Analysis (PCA)
Principal component analysis (PCA) is a classical method for reducing the number of attributes in the data by projecting the data from its original high-dimensional space into a lower-dimensional space.

The new attributes (also known as components) created by PCA have the following properties: (1) they are linear combinations of the original attributes, (2) they are orthogonal (perpendicular) to each other, and (3) they capture the maximum amount of variation in the data.

We use the sklearn PCA library to perform principal component analysis on our data.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
projected = pca.fit_transform(data2)
dataf = pd.DataFrame(projected,columns=['pc1','pc2','pc3'])
dataf.head()
```
